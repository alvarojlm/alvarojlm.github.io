```python
import cv2
import matplotlib.pyplot as plt
from numpy import array as array
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/PDI/img/pinguins.jpeg')
print(img.shape)
cv2_imshow(img)
```

```python
img_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
boost=array([[0, -1, 0], [-1, 5.2, -1], [0, -1, 0]])

imagemfiltra=cv2.filter2D(img_cinza,-1,boost)

cv2_imshow(img_cinza)
cv2_imshow(imagemfiltra)
```

```python
def laplaciano(img_cinza):
    lapl=array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])

    imagemfiltra=cv2.filter2D(img_cinza,-1,lapl)

    return imagemfiltra
def gaus(img_cinza):
    gaus=array([[0.0625, 0.125,  0.0625], [0.125, 0.25,
                   0.125],  [0.0625, 0.125,  0.0625]])

    imagemfiltra=cv2.filter2D(img_cinza,-1,gaus)

    return imagemfiltra
```

```python
img_lap=laplaciano(img_cinza)
cv2_imshow(img_lap)
```

```python
img_g=gaus(img_cinza)
img_lapG=laplaciano(img_g)
cv2_imshow(img_lapG)
```
