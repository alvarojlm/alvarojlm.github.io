```python
#@title Carregando Imagens com OpenCV
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/PDI/r.png") #imagem em P&B
plt.subplot(1,2,1) #primeiro plot
plt.imshow(img)

imcolor = cv2.imread('/content/drive/MyDrive/Colab Notebooks/PDI/mk.png') #imagem colorida lida em BRG
b,g,r = cv2.split(imcolor) #separa as matrizes de cor
img2 = cv2.merge([r,g,b]) #mistura as matrizes de cor em outra ordem
plt.subplot(1,2,2) #segundo plot
plt.imshow(img2)

plt.show()#mostrando os plots
```

```python
#@title Mudando um grupo de Pixels
import numpy as np

nova_cor = np.array([155,167,138]) #cor em BGR
img2=np.array(img) #nova imagem em array
plt.subplot(1,2,1)
plt.imshow(img)

for i in range(58, 355): #eixo y
  for j in range (170, 480): #eixo x
    img2[i][j]=nova_cor

plt.subplot(1,2,2)
plt.imshow(img2)

plt.xlabel('X')
plt.ylabel('y')
plt.rcParams['figure.figsize'] = [15, 15] #tamanho do plot
plt.show()
```
